```python
plt.clf()
sns.set_style('ticks')
sns.set_context('paper')
# Step 1: Group by phylum and calculate percentage of membrane-associated systems
phylum_stats = (
    df.groupby('phylum')['is_membrane_associated']
      .mean()  # mean gives fraction of True
      .reset_index()
)

# Step 2: Convert fraction to percentage
phylum_stats['percent_membrane_associated'] = phylum_stats['is_membrane_associated'] * 100

# Step 3: Filter out phyla with 0%
phylum_stats_nonzero = phylum_stats[phylum_stats['percent_membrane_associated'] > 0]

# Step 4: Plot
plt.figure(figsize=(10,6))
sns.barplot(
    data=phylum_stats_nonzero.sort_values('percent_membrane_associated', ascending=False),
    x='phylum',
    y='percent_membrane_associated',hue='phylum',
    palette='pastel',edgecolor='k',
)
plt.axhline((7916/ (164417+7916) )*100,linewidth=2,color='k',linestyle='--')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Percentage of Membrane-Associated\nAnti-Phage Defense Systems')
plt.xlabel('Phylum')
plt.savefig('Percentage of Membrane-Associated Anti-Phage Defense Systems.png',transparent=False,dpi=600)
plt.savefig('Percentage of Membrane-Associated Anti-Phage Defense Systems.svg',transparent=True,dpi=1000)
plt.tight_layout()

plt.show()
```

```python
# Step 1: Total number of defense systems per genome
total_defense_per_assembly = df.groupby('Assembly')['sys_id'].count().reset_index()
total_defense_per_assembly.rename(columns={'sys_id': 'total_defense_systems'}, inplace=True)

# Step 2: Number of membrane-associated defense systems per genome
membrane_defense_per_assembly = (
    df[df['is_membrane_associated'] == True]
    .groupby('Assembly')['sys_id']
    .count()
    .reset_index()
)
membrane_defense_per_assembly.rename(columns={'sys_id': 'membrane_defense_systems'}, inplace=True)

# Step 3: Merge both metrics
assembly_stats = total_defense_per_assembly.merge(membrane_defense_per_assembly, on='Assembly', how='left')
assembly_stats['membrane_defense_systems'] = assembly_stats['membrane_defense_systems'].fillna(0)

# Step 4: Plot
plt.figure(figsize=(15,6))
sns.scatterplot(
    data=assembly_stats,
    x='total_defense_systems',
    y='membrane_defense_systems',
    alpha=0.6
)
plt.xlabel('Total Defense Systems per Genome')
plt.ylabel('Membrane-Associated Defense Systems per Genome')
plt.title('Defense Systems per GenBank Assembly / Genome')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Prepare data (same as before)
total_defense_per_assembly = df.groupby('Assembly')['sys_id'].count().reset_index()
total_defense_per_assembly.rename(columns={'sys_id': 'total_defense_systems'}, inplace=True)

membrane_defense_per_assembly = (
    df[df['is_membrane_associated'] == True]
    .groupby('Assembly')['sys_id']
    .count()
    .reset_index()
)
membrane_defense_per_assembly.rename(columns={'sys_id': 'membrane_defense_systems'}, inplace=True)

assembly_stats = total_defense_per_assembly.merge(membrane_defense_per_assembly, on='Assembly', how='left')
assembly_stats['membrane_defense_systems'] = assembly_stats['membrane_defense_systems'].fillna(0)

# Step 2: Plot histogram with a bar per integer
plt.figure(figsize=(6,6))
sns.histplot(assembly_stats['total_defense_systems'], binwidth=1, color='skyblue', label='Total Defense Systems', kde=False, alpha=0.7)
sns.histplot(assembly_stats['membrane_defense_systems'], binwidth=1, color='salmon', label='Membrane Defense Systems', kde=False, alpha=0.7)

plt.xlabel('Number of Defense Systems per Genome')
plt.ylabel('Number of Genomes')
plt.title('Distribution of Defense Systems per GenBank Assembly / Genome (Integer Bins)')
plt.legend()
#plt.xticks(range(0, int(max(assembly_stats['total_defense_systems']))+1))
plt.tight_layout()
plt.xlim(0,30)
plt.show()
```

```python
plt.clf()
sns.set_style('ticks')
sns.set_context('paper')

# Step 1: Prepare data
total_defense_per_assembly = df.groupby('Assembly')['sys_id'].count().reset_index()
total_defense_per_assembly.rename(columns={'sys_id': 'total_defense_systems'}, inplace=True)

membrane_defense_per_assembly = (
    df[df['is_membrane_associated'] == True]
    .groupby('Assembly')['sys_id']
    .count()
    .reset_index()
)
membrane_defense_per_assembly.rename(columns={'sys_id': 'membrane_defense_systems'}, inplace=True)

assembly_stats = total_defense_per_assembly.merge(membrane_defense_per_assembly, on='Assembly', how='left')
assembly_stats['membrane_defense_systems'] = assembly_stats['membrane_defense_systems'].fillna(0)

# Step 2: Calculate non-membrane defense systems
assembly_stats['non_membrane_defense_systems'] = assembly_stats['total_defense_systems'] - assembly_stats['membrane_defense_systems']

# Step 3: Melt the dataframe for stacked plotting
stacked_data = assembly_stats.melt(
    id_vars=['Assembly'], 
    value_vars=['membrane_defense_systems','non_membrane_defense_systems',],
    var_name='defense_type', 
    value_name='count'
)

# Step 4: Plot stacked histogram
plt.figure(figsize=(4,4),dpi=300)
sns.histplot(
    data=stacked_data, 
    x='count', 
    hue='defense_type', 
    multiple='stack', 
    binwidth=1,
    palette={'membrane_defense_systems':'salmon', 'non_membrane_defense_systems':'skyblue'},
    edgecolor='black'
)
plt.xlabel('Number of Defense Systems per Genome')
plt.ylabel('Number of Genomes')
plt.xlim(0,30)
#plt.xticks(range(0, int(assembly_stats['total_defense_systems'].max())+1))
plt.legend(labels=[ 'Non-Membrane','Membrane',])
plt.tight_layout()
plt.savefig('Stacked Distribution of Defense Systems per Genome (Membrane vs Non-Membrane).png',transparent=False,dpi=600)
plt.show()
```

```python
plt.clf()
sns.set_style('ticks')
sns.set_context('paper')
# Step 1: Group by subtype and calculate fraction membrane-associated
subtype_stats = (
    df.groupby('subtype')['is_membrane_associated']
      .mean()  # mean of True/False gives fraction True
      .reset_index()
)

# Step 2: Convert fraction to percentage
subtype_stats['percent_membrane_associated'] = subtype_stats['is_membrane_associated'] * 100

# Step 3: Filter out subtypes with <5% membrane association
subtype_stats_filtered = subtype_stats[subtype_stats['percent_membrane_associated'] >= 5]

# Step 4: Plot
plt.figure(figsize=(12,6))
sns.barplot(
    data=subtype_stats_filtered.sort_values('percent_membrane_associated', ascending=False),
    x='subtype',
    y='percent_membrane_associated',
    palette='magma'
)
plt.xticks(rotation=45, ha='right')
plt.ylabel('% Membrane-Associated')
plt.xlabel('Defense System Subtype')
plt.title('Percentage of Membrane-Associated Systems by Subtype (≥5%)')
plt.tight_layout()
plt.show()
```
